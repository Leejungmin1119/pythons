```python
"""
LIS 알고리즘 (최장 증가 수열)

원소를 n개 가지고 있는 배열에서 일부를 골라내어 만든 수열이 오름차순으로 정렬되어있으면서 길이가 가장 긴 수열을 말함.


1. LIS 배열을 생성

2. LIS 배열의 마지막 원소값이 현 비교할려는 원소값보다 크면 LIS 배열 뒤에 값을 저장

3. 만약 비교할려는 원소값이 작다면 이분탐색을 통해서 들어갈 위치에 저장

예시
1. 30 40 45 10 20

# 값이 더 크니까 저장
{30}
{30,40}
{30,40,45}

# 45 > 10 이니깐 이진 탐색을 통해서 해당 값이 있어야 하는 위치를 찾아 교체, 교체를 했다고 해서 최장 길이를 못찾는 건 아니지만.
#  정확한 LIS 의 수열의 값들을 파악할 수는 없다.

{10,40,45}
{10,20,45} 값 = 3

2.10 20 10 25 21 24

{10}
{10,20}
{10,20}
{10,20,25}
{10,20,21}
{10,20,21,24} 값 = 4 
{10,30,40,50,25}

1. 30 40 45 10 20

{30,40,45} 10

30 40 45 31 20

{30 31 45} 이 정답 하지만 {31,40,45} 이렇게 되면 안됨. 
"""
import sys
# 이분탐색
def binary_search(start,end,target):


    while start <= end:

        mid = int((start+end)/2)

        # 만약 찾는 값이 없고 아래조건처럼 mid보단 큰데 mid+1 보단 작으면 mid+1 에 저장
        if LIS[mid] < lists[target] < LIS[mid+1]:
            LIS[mid+1] = lists[target]
            return 0
        
        elif LIS[mid] == lists[target]:
            LIS[mid] = lists[target]
            return 0
        elif LIS[mid] < lists[target]:
            start = mid+1
        elif LIS[mid] > lists[target]:
            end = mid-1
        
    LIS[mid] = lists[target]
    return 0




# 메인함수

num = int(sys.stdin.readline())

lists = list(map(int,sys.stdin.readline().split()))
LIS = []
for i in range(num):

    # 만약 LIS 의 마지막 값보다 비교할려는 원소값이 더 크면 바로 저장
    if not LIS or LIS[-1] < lists[i]:
        LIS.append(lists[i])

    # 만약 LIS 의 마지막 값보다 비교할려는 원소값이 더 작으면 이분 탐색을 통하여 들어갈 위치를 찾는다.
    elif LIS[-1] >= lists[i]:

        binary_search(0,len(LIS)-1,i)

print(len(LIS))

```
