```python
#https://www.acmicpc.net/problem/1918
# 백준 알고리즘 1918번 후위 표기식 (골드 2)

"""
후위 표기식 (스택 알고리즘)

1. 제일 처음 알파벳들은 알파벳끼리 스택에 저장한다.
2. 알파벳을 스택에 저장하다가 만약 기호들을 만난다면, 그 기호들만 모이는 곳에 저장한다.
3. 기호 스택에 기호가 있는 상태에서 기호가 오게 된다면 다음에 따라서 행동한다.

? 스택 안에 있는 연산자보다 우선순위가 높은가?
- 저장하기

? 스택 안에 있는 연산자 보다 우선순위가 낮은가?
- 스택안에 있는 기호들을 전부 알파벳이 있는 스택에 저장, 그리고
새로운 기호를 다시 저장한다.

4. 연산이 다 끝나면 기호 스택 안에 있는 나머지 기호들을 다 알파벳 기호에다가 저장하고,
출력한다.




주위!

3번에서 우선순위를 판단할떄 무조건 +,-,* 이순으로 보는 거 뿐만 아니라 괄호가 처져 있는 경우가 있으므로

그것또한 봐주면서 우선순위를 판단해야 한다.


"""
#이동 및 새롭게 저장
def stack_move(a):
    global ch_num,ch_symbol
    for j in range(len(stack_bracket)-1,-1,-1):
        
        if (stack_bracket[j] == '('):
            stack_bracket.pop()
            continue
        
        stack_string.append(stack_bracket[j])
            
    # 기존 기호 스택은 비우고 다시 새로운 기호 저장                 
    stack_bracket.clear()
    stack_bracket.append(a)
    
    
    
string = list(input())
stack_string =[]
stack_bracket = [] # [기호,괄호순,우선순위]


ch_symbol = '+'

#문자열 확인
for i in string:
    
    if stack_bracket:
        ch_symbol = stack_bracket[len(stack_bracket)-1][0]
        
    # 만약 해당 부분이 괄호라면 check++  or check--(우선순위 높이기)
    if i == "(":
        stack_bracket.append(i)
    elif i == ")":
        for j in range(len(stack_bracket)-1,-1,-1):
            
            if (stack_bracket[j] == '('):
                stack_bracket.pop()
                break
            
            stack_string.append(stack_bracket[j])
            
    #기본적인 연산중 가장 낮은 순위
    elif (i == "+" or i == "-") and (ch_symbol == '+' or ch_symbol == '-'):  
        stack_bracket.append(i)

    # 높은 순위
    elif i == "*" or i == "/":
        stack_bracket.append(i)
    
    #알파벳       
    elif i not in ["+","-","*","/"]:
        stack_string.append(i)
        
    # 위에 해당하지 않으면 옮기기
    else:
        stack_move(i)

#남아 있는 것들 옮기기
for i in range(len(stack_bracket)-1,-1,-1):
    stack_string.append(stack_bracket[i])

# 출력
for i in stack_string:
    
    print(i,end="")
    
    
```
