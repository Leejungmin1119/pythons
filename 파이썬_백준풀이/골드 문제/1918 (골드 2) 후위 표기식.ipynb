```python
#https://www.acmicpc.net/problem/1918
# 백준 알고리즘 1918번 후위 표기식 (골드 2)

"""
후위 표기식 (스택 알고리즘)

1. 제일 처음 알파벳들은 알파벳끼리 스택에 저장한다.
2. 알파벳을 스택에 저장하다가 만약 기호들을 만난다면, 그 기호들만 모이는 곳에 저장한다.
3. 기호 스택에 기호가 있는 상태에서 기호가 오게 된다면 다음에 따라서 행동한다.

? 스택 안에 있는 연산자보다 우선순위가 높은가?
- 저장하기

? 스택 안에 있는 연산자 보다 우선순위가 낮은가?
- 스택안에 있는 기호들을 전부 알파벳이 있는 스택에 저장, 그리고
새로운 기호를 다시 저장한다.

4. 연산이 다 끝나면 기호 스택 안에 있는 나머지 기호들을 다 알파벳 기호에다가 저장하고,
출력한다.




주위!

3번에서 우선순위를 판단할떄 무조건 +,-,* 이순으로 보는 거 뿐만 아니라 괄호가 처져 있는 경우가 있으므로

그것또한 봐주면서 우선순위를 판단해야 한다.


"""

#나중에 풀기 

#이동 및 새롭게 저장



String = list(input())

# 기호를 저장할 리스트와 문자열 선언
ans = ""#후위 표기식
symbol = []#기호저장 리스트

for i in String:
    # isalpha == 알파벳인지 아닌지 확인해준다
    if i.isalpha():
        ans+=i
        
    # 만약 +라면? 괄호가 오지 않는 이상 우선순위는 높을 수가 없으므로 pop을 한다.
    elif i == '+' or i == '-':
        while symbol and symbol[-1] != "(":
            ans+=symbol.pop()
        symbol.append(i)
            
    elif i == '/' or i == '*':
        # 만약 이전의 값의 / 나 * 면 우선순위가 같으므로 pop , 그리고 symbol[-1] 의값이 ( 라면 우선순위가 높다는 뜻이므로 pop을 하지 않는다.
        while (symbol) and (symbol[-1] == "/" or symbol[-1] == "*"):
            ans+=symbol.pop()
        symbol.append(i)
        
    # (라면 기호 스택에 추가한다. 이는 우선순위 파악 및 어디까지 pop해야하는지 알수 있는 역할이 된다.     
    elif i == "(":
        symbol.append(i)
    
    elif i == ')':
        while symbol and symbol[-1] != "(":
            ans+=symbol.pop()
        symbol.pop()
            
            
while symbol:
    ans+=symbol.pop()
#출력
for i in range(len(ans)):

    print(ans[i],end="")

                














```
