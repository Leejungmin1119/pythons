```python
# 적록색약 : 빨강 - 초록 = 같은색깔로 취급

"""
총 4가지의 경우의 수 가 있다.

1. 빨강으로 탐색

2. 초록으로 탐색

3. 빨-초 같은색으로 하고 탐색

4. 파랑으로 탐색 (적록색약과 상관없이 파랑의 갯수는 같다.)

총 4가지의 수로 탐색을 하여 가지의 수를 찾아 출력하면 되는 문제이다.

"""
from collections import deque
import sys
def BFS(y,x,pass_num,*color):

    # 적록색약인지 사람인지 구분
    passes = [0,1]

    global num
    #튜플의 형태로 저장
    deques = deque([[y,x]])


    while deques:

        y,x = deques.popleft()

        #자기자신 방문처리
        ch[y][x] = pass_num

        for i in range(4):
            ch_y,ch_x = move[i][0]+y,move[i][1] + x

            # 갈려는 방향이 인덱스를 벗어놨는지 확인
            if  0<= ch_y < num and 0 <= ch_x < num :

                # 이미 간 방향인지 확인
                if ch[ch_y][ch_x] == passes[pass_num-1] and lists[ch_y][ch_x] in color:

                    # 방문 처리
                    ch[ch_y][ch_x] = pass_num
                    deques.append((ch_y,ch_x))

# 메인함수
num = int(sys.stdin.readline())

#방향 탐색 리스트
move = [[0,1],[0,-1],[1,0],[-1,0]]

lists = [list(sys.stdin.readline()) for i in range(num)]

ch = [[0] * num for i in range(num)]

dic = {"R":0,"G":0,"B":0}

#일반인 기준으로 한번 탐색
for i in range(num):
    for j in range(num):

        if ch[i][j] == 0:\
            #딕셔너리에 방문 추가
            dic[lists[i][j]]+=1
            BFS(i,j,1,lists[i][j])

people = sum(dic.values())

#초기화
dic["R"],dic["G"] = 0,0
#적록색약을 가지고 있는 조건에서 탐색

for i in range(num):
    for j in range(num):
        
        # 이미 파랑은 구했기 떄문에 불필요하게 탐색 필요 x
        if ch[i][j] == 1 and lists[i][j] != "B":
            dic[lists[i][j]]+=1
            BFS(i,j,0,"R","G")

same_people = sum(dic.values())

print(f"{people} {same_people}")
```
