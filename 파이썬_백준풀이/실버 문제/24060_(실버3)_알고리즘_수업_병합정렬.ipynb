```python
# https://www.acmicpc.net/problem/24060
# 24060 알고리즘 수업 병합정렬 (실버3) / 병합정렬, 재귀

"""
병합 정렬 : 문제를 작은 2문제로 나눈 후 다시 합치면서 정렬하는 방식
- 무조건 재귀 함수를 사용해야함 - 

1. 리스트를 반으로 나누어서 리스트 크기가 1이 되도록 만드는 재귀함수 구현 

2. 분할된 리스트를 새로운 리스트에 정렬하여 저장 하는 함수 구현

3. 저장할때 몇번째인지 구별하고 또한 해당 저장된 수가 어떤거인지 저장하는 변수 구현 만약 총 저장 수가 k보다 적으면 -1 출혁
"""
#분할 정렬 알고리즘
def merge_sorted(lists,start,end):

    mid = int((start+end)/2)# 리스트를 반으로 나누기 위한 mid 값 구하기

    if start < end:
        #분할
        merge_sorted(lists,start,mid)# 전처리
        merge_sorted(lists,mid+1,end)# 후처리

        #합병
        merge(lists,start,mid,end)# 분할을 다하면 마지막으로 합병 시작 
#합병함수
def merge(lists,start,mid,end):
    global merge_list
    global count, save_num
    merge_list = []

    ch_low, ch_high = start,mid+1

    # 정렬 끝났는지 확인
    while ch_low <= mid and ch_high <= end:

    # 리스트 0 와 mid +1 번째 리스트의 값끼리 비교
        if lists[ch_low] >= lists[ch_high]:
            merge_list.append(lists[ch_high])
            
            ch_high = ch_high + 1
        else:
            merge_list.append(lists[ch_low])
            ch_low = ch_low + 1

        count = count+1#저장 횟수 기록

        # 원하는 장소에 찾으면 출력
        if count == save_num:
            print(merge_list[-1])

    # 왼쪽밖에 안남았을때
    while ch_low <= mid:
        merge_list.append(lists[ch_low])
        ch_low = ch_low + 1

        count = count+1

        if count == save_num:
            print(merge_list[-1])

    # 오른쪽밖에 안남았을때
    while ch_high <= end:
        merge_list.append(lists[ch_high])
        ch_high = ch_high + 1

        count = count+1

        if count == save_num:
            print(merge_list[-1])
        

    #재저장
    lists[start:end+1] = merge_list[0:len(merge_list)]
#메인

# 병합정렬 할때 쓰일 전역 변수 선언
merge_list = []

count = 0
list_num, save_num = map(int,input().split())

lists = list(map(int,input().split()))

merge_sorted(lists,0,len(lists)-1)

if save_num > count:
    print(-1)

```
