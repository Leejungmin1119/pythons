```python
#https://www.acmicpc.net/problem/24445
# 24445번 알고리즘 수업 너비 우선탐색2 (실버 2)
import sys
from collections import deque
sys.setrecursionlimit(10**6)
def BFS(start):
    global count
    # 큐 선언
    queue = deque([start])

    node = set()
    while queue:
        ch = queue.popleft()
        #[n,n] 방지
        if ch not in node:
            node.add(ch)
            #순서 정보 저장
            ch_count[ch-1] = count
            count+=1

            # 내가 갈려도 곳이 이미 방문한 곳인지 확인
            for i in dic[ch]:
                #만약 방문안한 곳이라면 deque에 추가
                if i not in node:    
                    queue.append(i)

count = 1
dot,line,num = map(int,sys.stdin.readline().split())

ch_count = [0]*dot
# dictionory 선언
dic = {i:[0] for i in range(1,dot+1)}


for i in range(line):
    line1,line2 = map(int,sys.stdin.readline().split())
# 입력 후 바로 값 바로 추가
    dic[line1].append(line2)
    dic[line2].append(line1)


for i in dic:
    #정렬
    dic[i].remove(0)
    dic[i].sort(reverse= True)

#실헹
BFS(num)

for i in ch_count:
    print(i)
```
