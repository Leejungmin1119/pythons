```python
#https://www.acmicpc.net/problem/24480
#24480 알고리즘 수업 깊이 우선탐색 (실버 2)


"""
1. input 후 간선 정보 딕셔너리에 저장

2. 딕셔너리에 저장된 내용들을 정렬 후 함수 실행

3. DFS 알고리즘 이용(재귀)

4. 출력

#입력
a,b,c = input
lists =  [::] #딕셔너리를 미리 선언하여 후반 넣기 편하게 설정

n1 n2 = input
lists = [n1,n2] # 딕셔너리에 바로 정보 넣기

lists.sort() # 정보 정렬 (역순)

dict()

"""
import sys

sys.setrecursionlimit(10**6)
# DFS 함수
def DFS(start):
    global count

    #[0,0,0,0,0,0] / ch_count[위치] = 순서 값
    ch_count[start-1] = count
    count+=1

    for i in dic[start]:
        # ch_countt[i-1] == 0 즉 내가 갈려고 하는 노드의 해당 값이 0이면 방문을 안한 것이고 0이 아니면 이미 방문한 노드이다. 
        if ch_count[i-1] == 0:
            #재귀 실행
            DFS(i)

#메인
count = 1
dot,line,num = map(int,sys.stdin.readline().split())

ch_count = [0]*dot
# dictionory 선언
dic = {i:[0] for i in range(1,dot+1)}


for i in range(line):
    line1,line2 = map(int,sys.stdin.readline().split())
# 입력 후 바로 값 바로 추가
    dic[line1].append(line2)
    dic[line2].append(line1)


for i in dic:
    #정렬
    dic[i].remove(0)
    dic[i].sort(reverse = True)

#실헹
DFS(num)

#결과
for i in ch_count:
    print(i)
```
